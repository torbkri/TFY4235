```python
import numpy as np
from task3_1 import *
```

```python
T_c = 2 / np.log(1 + np.sqrt(2)) 
def m_vs_ave(T = T_c):
    L = 40
    n_sweeps = 10_000_000
    
    E_evolution = np.zeros(n_sweeps)
    M_evolution = np.zeros(n_sweeps)
    n_equilibrium = 2000
    
    # equilibrate:
    spins, _ , _ = run_metropolis(L, n_equilibrium, T)

    # run the simulation:

    _, spin_sums, total_Es = run_metropolis(L, n_sweeps, T, spins=spins)
    
    sample_averages = np.array([1_000, 10_000, 100_000, 1_000_000, 10_000_000])
    M_values = np.zeros(len(sample_averages))
    E_evolution = np.zeros(len(sample_averages))
    M_var = np.zeros(len(sample_averages))
    E_var = np.zeros(len(sample_averages))
    for i in range(len(sample_averages)):
        M_values[i] = np.mean(spin_sums[:sample_averages[i]])
        E_evolution[i] = np.mean(total_Es[:sample_averages[i]])
        M_var[i] = np.var(spin_sums[:sample_averages[i]])
        E_var[i] = np.var(total_Es[:sample_averages[i]])

    np.savez(f'convergence.npz', M_values=M_values, E_evolution=E_evolution, M_var=M_var, E_var=E_var, sample_averages=sample_averages)    
```

```python
m_vs_ave()
```
