```python
plt.rcParams.update({
    "text.usetex": True,  
    "font.family": "serif",
    "font.serif": ["Times"], 
})

def black_text(ax, text, xfrac, yfrac, fontsize=30):
    ax.annotate(
        text, xy=(xfrac, yfrac), fontsize=fontsize, color='black',
        xycoords='axes fraction'
    )
```

# 3.1(1)

```python
labelsize = 26
ticksize = 22
legendsize = 20
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 5))


data = np.load('M_time_nofield.npz')
Ts = data['Ts']
spin_sums = data['spin_sums']
print(spin_sums)

for i in range(len(Ts)):
    T = Ts[i]
    mag = spin_sums[i]
    steps = np.arange(len(mag))
    ax.scatter(steps, mag, label=f"T = {T}", s = 1)
ax.set_ylim(-1, 1)
```

```python
ax.legend(fontsize=legendsize)
ax.legend(scatterpoints=1, markerscale=10, fontsize=legendsize, ncol=3, columnspacing=0.01)
ax.tick_params(axis='x', labelsize=ticksize)
ax.tick_params(axis='y', labelsize=ticksize)
ax.set_xlabel(r'\# sweeps', fontsize=labelsize)
ax.set_ylabel(r'M', fontsize=labelsize)
```

```python
fig.tight_layout()
```

# 3.2(3)

```python
data = np.load('scaling_field2.npz')
T_range = data['T_range']
chi_values = data['chi_scaled_vals']
h_values = data['h_scaled_vals']
H_values = data['H_list']
```

```python
# version 2, må skalere chi med L^2 av samme grunn som før

fig, ax = plt.subplots(figsize=(15, 5))
T_c = 2 / (np.log(1 + np.sqrt(2)))

for i in range(len(H_values)):
    H = H_values[i]
    below = T_range < T_c
    ax.scatter(h_values[i, below], chi_values[i, below]*40**2, s=5, label = f'H  = {H}, '+ r'$T < T_c$')

for i in range(len(H_values)):
    H = H_values[i]
    above = T_range > T_c
    ax.scatter(h_values[i, above], chi_values[i, above]*40**2, s=5, label = f'H = {H}, '+ r'$T > T_c$') 


ax.legend(scatterpoints=1, markerscale=5, fontsize=legendsize, ncols = 2, columnspacing=0.01)
ax.set_xscale('log')
ax.set_yscale('log')
```

```python
ax.tick_params(axis='x', labelsize=ticksize)
ax.tick_params(axis='y', labelsize=ticksize)
ax.set_xlabel(r'$h / |t|^{\beta \delta}$', fontsize=labelsize)
ax.set_ylabel(r'$\chi \, |t|^{\gamma}$', fontsize=labelsize)
```

```python
fig.tight_layout()
```
