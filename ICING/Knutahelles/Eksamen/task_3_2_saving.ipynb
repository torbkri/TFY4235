# Finding susceptibility with different H

```python
def task_2():

    H_values = np.array([0.01, 0.03, 0.05])
    L = 40
    n_sweeps = 12000
    T_min = 2
    T_max = 4
    T_res = 400
    ave = 10000
    chi_values = np.zeros((len(H_values), T_res))
    M_values = np.zeros((len(H_values), T_res))


    for i in range(len(H_values)):
        H = H_values[i]
        T_range, chi_value, M_value = find_susceptiblity(L = L, ave=ave, n_sweeps=n_sweeps, H = H, T_min=T_min, T_max=T_max, T_res=T_res)
        chi_values[i] = chi_value
        M_values[i] = M_value

    np.savez('susceptibility_field2.npz', T_range=T_range, chi_values=chi_values, M_values=M_values, H_values=H_values)    
```

```python
task_2()
```

# Investigating data collapse

```python
def task_3():
    gamma, beta, delta = 7/4, 1/8, 15
    T_c = 2 / (np.log(1 + np.sqrt(2)))

    L, n_sweeps, ave = 40, 12000, 10000
    T_min, T_max, T_res = 1.0, 4.0, 400

    H_list = np.array([0.01, 0.03, 0.05])

    chi_scaled_vals = np.zeros((len(H_list), T_res))
    h_scaled_vals = np.zeros((len(H_list), T_res))

    for i in range(len(H_list)):
        H = H_list[i]

        T_range, chi_values, _ = find_susceptiblity(
            L, ave, n_sweeps, T_min, T_max, T_res, H
        )

        t_abs = np.abs((T_range - T_c) / T_c)
        chi_scaled = chi_values * (t_abs**gamma)
        h_scaled   = H / t_abs**(beta * delta)
        chi_scaled_vals[i] = chi_scaled
        h_scaled_vals[i] = h_scaled

    np.savez(
        'scaling_field2.npz',
        T_range=T_range,
        chi_scaled_vals=chi_scaled_vals,
        h_scaled_vals=h_scaled_vals,
        H_list=H_list,
    )
```
