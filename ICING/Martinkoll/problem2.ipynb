```python
import numpy as np
```

```python
#Problem 2a
x = 10**8
y = 1.0


z_half = (np.float16(x) + np.float16(y)) - np.float16(x)
print("Result with half:", z_half)

z_single = (np.float32(x) + np.float32(y)) - np.float32(x)
print("Result with single:", z_single)

z_double = (np.float64(x) + np.float64(y)) - np.float64(x)
print("Result with double:", z_double)
```

```python
#Problem 2b

a = 0.1 + 0.2
b = 0.3
print("a == b:", a == b)
print("a", a, "b", b)
print(abs(a-b) < 1e-15)
```

```python
#Problem 2c

x = 1.0
y = 1.0 + 10**6 * 10**-16

increment = 10**-16
steps = 10**6

for i in range(steps):
    x += increment

# Print results
print(f"x: {x:.20f}")
print(f"y: {y:.20f}")
print(f"Difference (x - y): {x - y:.20f}")

```

```python
#Problem 2d

#float64 is double precision
factorial_170 = np.float64(1)

for i in range(1, 171):  
    factorial_170 *= np.float64(i)

factorial_171 = factorial_170 * np.float64(171)

print(f"170! double precision = {factorial_170}")
print(f"171! double precision = {factorial_171}")
```

```python
#Problem 2e

a = np.float64(1.000000000001)
b = np.float64(1.0)

unstable64 = np.sqrt(a) - np.sqrt(b)

stable64 = (a - b) / (np.sqrt(a) + np.sqrt(b))

print(f"Unstable64: {unstable64:.20e}")
print(f"Stable64  : {stable64:.20e}")

print(f"Absolute difference: {abs(unstable64 - stable64):.2e}")


```
