```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def get_filter(shape, radius):
    H, W = shape
    y = np.arange(H)
    x = np.arange(W)
    
    yy, xx = np.meshgrid(y, x, indexing='ij')
    
    cy = H // 2
    cx = W // 2
    
    return (yy - cy)**2 + (xx - cx)**2 <= radius**2
```

```python
img = cv2.imread("/Users/torbjornkringeland/Desktop/NTNU/6/TFY4235/SFD-STUDENTS/exampleSignature.png", cv2.IMREAD_GRAYSCALE)
img
```

```python
ft_img = np.fft.fft2(img)
# ft_img = np.abs(ft_img)

ft_img.max()
centered = np.fft.fftshift(ft_img)


# result =np.abs(np.fft.ifft2(filtered))
```

```python
filter = get_filter(ft_img.shape, 50)

filtered = np.zeros((ft_img.shape[0],ft_img.shape[1]),dtype= int)
for y in range(filtered.shape[0]):
    for x in range(filtered.shape[1]):
        filtered[y][x] = filter[y][x] * (centered[y][x])
```

```python
result = np.fft.ifft2(filtered)
plt.imshow(np.abs(result))
```

```python
# plt.hist(np.log(ft_img.ravel()), bins = 100)
fig,ax = plt.subplots(2,2, figsize = (8,8))
ax[0,0].hist(np.log(ft_img.ravel()), bins = 100)
ax[0,1].imshow(np.log(np.abs(centered)),cmap = "Grays")
ax[1,0].imshow(np.log(np.abs(filtered)), cmap="Grays")
ax[1,1].imshow(np.abs(result),cmap = "Grays")
fig.tight_layout()
```
