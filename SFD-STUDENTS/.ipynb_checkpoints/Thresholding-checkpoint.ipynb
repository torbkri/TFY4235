```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

# from Convolution import display_img
```

```python
def display_img(image):
    fig, ax = plt.subplots(1, 2, figsize=(10, 5))
    ax[0].imshow(image, cmap='gray')
    ax[0].axis('off')
    ax[0].set_title('Image')
    ax[1].hist(image.ravel(), bins=256, range=(0, 256))
    ax[1].set_title('Histogram')
    ax[1].set_xlabel('Pixel Intensity')
    ax[1].set_ylabel('Frequency (log scale)')
    ax[1].set_yscale('log')
    fig.tight_layout()
    plt.show()
    plt.close()
```

```python
def rotate(image, theta) -> np.ndarray:
    rotation_matrix = cv2.getRotationMatrix2D((image.shape[1]//2, image.shape[0]//2), theta, 1.0)
    rotated = cv2.warpAffine(image, M = rotation_matrix, dsize=(image.shape[1], image.shape[0]))
    return rotated
```

```python
rotated = rotate(img, -40)
```

```python
plt.imshow(rotated)
```
