```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/Users/torbjornkringeland/Desktop/NTNU/6/TFY4235/SFD-STUDENTS/exampleSignature.png', cv2.IMREAD_GRAYSCALE)
print(img.shape)
plt.imshow(img, cmap='gray')
```

```python
def display_img(image):
    fig, ax = plt.subplots(1, 2, figsize=(10, 5))
    ax[0].imshow(image, cmap='gray')
    ax[0].axis('off')
    ax[0].set_title('Image')
    ax[1].hist(image.ravel(), bins=256, range=(0, 256))
    ax[1].set_title('Histogram')
    ax[1].set_xlabel('Pixel Intensity')
    ax[1].set_ylabel('Frequency (log scale)')
    ax[1].set_yscale('log')
    fig.tight_layout()
    plt.show()
    plt.close()
```

```python
kernel_blur = np.array([[1, 1, 1],
                        [1, 1, 1],
                        [1, 1, 1]]) / 9.0
kernel_sharp = np.array([[0, -5, 0],
                        [-5, 25, -5],
                        [0, -5, 0]])
kernel_B = np.array([[0,-1,0],
                     [-1,5,-1],
                     [0,-1,0]])
kernel_C = np.array([[-1,0,1],
                     [-2,0,2],
                     [-1,0,1]])
kernel_gaussian:np.ndarray = np.array([[1, 2, 1],
                            [2, 4, 2],
                            [1, 2, 1]]) / 16.0
blurred_img = cv2.filter2D(img, -1, kernel_blur)
sharpened_img = cv2.filter2D(img, -1, kernel_sharp)
gaussian_img = cv2.filter2D(img, -1, kernel_gaussian)
fig, ax = plt.subplots(2, 2, figsize=(10, 10))
ax[0, 0].imshow(img, cmap='gray')
ax[0, 0].set_title('Original Image'); ax[0, 0].set_axis_off()
ax[0, 1].imshow(blurred_img, cmap='gray')
ax[0, 1].set_title('Blurred Image'); ax[0, 1].set_axis_off()
ax[1, 0].imshow(sharpened_img, cmap='gray')
ax[1, 0].set_title('Sharpened Image'); ax[1, 0].set_axis_off()
ax[1, 1].imshow(gaussian_img, cmap='gray')
ax[1, 1].set_title('Gaussian Image'); ax[1, 1].set_axis_off()
```

```python

B_image = cv2.filter2D(cv2.medianBlur(img,5), -1, kernel_B)
display_img(B_image)

#Kernel B smears edges and evens out written parts
```

```python

C_image = cv2.filter2D(cv2.medianBlur(img,5), -1, kernel_C)
display_img(C_image)

#Kernel C removes parts not on the right edge of writing. In other words it gives the shadow of shapes.
```
