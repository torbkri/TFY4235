# Signature Verification with TensorFlow/Keras (Same-Identity Pairs)
This notebook provides a lean, readable baseline to load same-identity pairs (genuine–genuine, genuine–forged) via `tf.data` and a minimal Keras model.

```python
import os, re, random, math
from pathlib import Path
from collections import defaultdict
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
import cv2

SEED = 42
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

# Global canvas size used across dataset and model
CANVAS_SIZE = (375, 616)

# Enable GPU memory growth (if GPU present)
gpus = tf.config.list_physical_devices('GPU')
for g in gpus:
    try: tf.config.experimental.set_memory_growth(g, True)
    except Exception: pass
    
DATA_ROOT = Path(os.path.abspath(os.getcwd()).rsplit('/',1)[0])
# DF_DIR = Path("./signatures-dataset/filtered_dataframe.csv")
DF_DIR_BIN = Path(f"{DATA_ROOT}/signatures-dataset/filtered_dataframe_binarized.csv")
# if 'df' not in globals():
#     df = pd.read_csv(DF_DIR)
if 'df_bin' not in globals():
    df_bin = pd.read_csv(DF_DIR_BIN)

df_bin["image_path_filtered"] = df_bin["image_path_filtered"].apply(
    lambda p: str((DATA_ROOT / Path(str(p).replace("\\", "/"))).resolve())
)
```

### Combinatorics

Let $P_{gf}^i = (S_g^i, S_f^i)$ be a pair of a genuine signature $S_g^i$ and a forged signature $S_f^i$ pertaining to the same identity $i$, and let $P_{gg}^i$ be a pair of two genuine and unique signatures pertaining to the same identity. We want to build a dataset of $P_{gf}^i$ and $P_{gg}^i$, where pairs are treated as features and targets are encoded as binary labels:

\begin{equation}
\begin{aligned}
\Bigg[
\text{Features} =
\begin{cases}
P_{gg}^i \\
P_{gf}^i \
\end{cases}
,\quad
\text{Target} =
\begin{cases}
0 & gg \\
1 & gf
\end{cases}
\end{aligned}
\Bigg]
\end{equation}

Let $N_g^i$ and $N_f^i$ be the number of genuine signatures and forged signatures respectively for identity $i$. From our preprocessing we ensured equal number of signatures for per person $i$, and equal number of genuine and forged signatures:

\begin{equation}
N_g^i = N_f^i =  N_g = N_f = N
\end{equation}

We can then generate $(2N)^2$ pairs, of which only half of these are eligable because the first signature must be genuine ($P_{gg}$ or $P_{gf}$), and in which $N$ of these eligable pairs are indistinguishible since both signatures are identical. For each person $i$ we therefore get the following number of eligable pairs $N_p^i$:

\begin{equation}
N_p^i = N_p = \frac{(2N)^2}{2}-N = N(2N - 1)
\end{equation}

We can also find this formula by thinking that we first chose one genuine pair; $N \text{C}1$, and then chose one more signature from the remaining signatures; $(2N-1)\text{C}1$, which multiplies to:

\begin{equation}
N \text{C}1 \times (2N-1)\text{C}1 = N(2N-1)
\end{equation}

```python
def get_pair_from_df(df, N=4, Np=32, seed=SEED):
    rng = random.Random(seed)

    pairs = defaultdict(list)
    identities = df["person_id"].unique()
    assert len(identities) == Np, f"Dataframe has {len(identities)} identities, expected {Np}"

    for person in identities:
        person_df = df[df["person_id"] == person]
        assert len(person_df) == 2 * N, f"Person {person} has {len(person_df)} rows, expected {2 * N}"
        assert person_df["real"].sum() == N, f"Person {person} does not have {N} genuine signatures"

        genuine_signatures = person_df[person_df["real"] == 1]["image_path_filtered"]
        forged_signatures = person_df[person_df["real"] == 0]["image_path_filtered"]

        for i, genuine in enumerate(genuine_signatures):
            other_genuines = [g for j, g in enumerate(genuine_signatures) if j != i]
            for other in other_genuines:
                pairs[person].append((genuine, other, 0))

        for genuine in genuine_signatures:
            for forged in forged_signatures:
                pairs[person].append((genuine, forged, 1))

        rng.shuffle(pairs[person])

    for person, person_pairs in pairs.items():
        assert len(person_pairs) == N * (2 * N - 1), f"Person {person} has {len(person_pairs)} pairs, expected {N * (2 * N - 1)}"

    assert len(pairs) == Np, f"Generated pairs for {len(pairs)} identities, expected {Np}"
    return pairs

# pairs = get_pair_from_df(df, N=4, Np=32, seed=SEED)
pairs_bin = get_pair_from_df(df_bin, N=4, Np=32, seed=SEED)
```

```python
def make_tf_dataset_from_pairs(pairs, batch_size=16, shuffle=True, seed=SEED,
                               image_shape=(CANVAS_SIZE[0], CANVAS_SIZE[1], 1)):
    """
    Create a tf.data.Dataset from generated pairs where x1/x2 are image paths.
    """
    def _load_and_preprocess(path):
        img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
        assert img is not None, f"Failed to load image: {path}"

        # Assert value range strictly in [0,1]
        minv, maxv = img.min(), img.max()
        if minv >= 0 and maxv <= 1:
            img = img.astype(np.float32)
        else:
            img = img.astype(np.float32) / 255.0
            assert img.min() >= 0.0 and img.max() <= 1.0, f"Image not in [0,1]: {path}"

        # Assert exact shape
        h, w = img.shape[:2]
        expected_h, expected_w = CANVAS_SIZE
        assert (h, w) == (expected_h, expected_w), f"Wrong size {img.shape} for {path}, expected {(expected_h, expected_w)}"

        # Convert to tensor (H, W, 1) for Conv2D
        img = tf.convert_to_tensor(img, dtype=tf.float32)
        img = tf.expand_dims(img, axis=-1)
        return img

    def pair_generator():
        for person_pairs in pairs.values():
            for x1_path, x2_path, label in person_pairs:
                x1 = _load_and_preprocess(x1_path)
                x2 = _load_and_preprocess(x2_path)
                y  = tf.convert_to_tensor(label, dtype=tf.float32)
                # Yield ((inputs...), label)
                yield ((x1, x2), y)

    output_signature = (
        (
            tf.TensorSpec(shape=image_shape, dtype=tf.float32),  # Image 1
            tf.TensorSpec(shape=image_shape, dtype=tf.float32),  # Image 2
        ),
        tf.TensorSpec(shape=(), dtype=tf.float32),               # Label
    )

    ds = tf.data.Dataset.from_generator(pair_generator, output_signature=output_signature)
    if shuffle:
        ds = ds.shuffle(buffer_size=1000, seed=seed, reshuffle_each_iteration=True)
    ds = ds.batch(batch_size).prefetch(tf.data.AUTOTUNE)
    return ds

def split_person_ids(eligible_ids, train_ratio=0.7, val_ratio=0.15, seed=SEED):
    """
    Split person_ids into train, validation, and test sets.

    Args:
        eligible_ids (list): List of person_ids.
        train_ratio (float): Proportion of person_ids for training.
        val_ratio (float): Proportion of person_ids for validation.
        seed (int): Random seed for reproducibility.

    Returns:
        tuple: (train_ids, val_ids, test_ids)
    """
    rng = random.Random(seed)
    rng.shuffle(eligible_ids)

    n_total = len(eligible_ids)
    n_train = int(n_total * train_ratio)
    n_val = int(n_total * val_ratio)

    train_ids = eligible_ids[:n_train]
    val_ids = eligible_ids[n_train:n_train + n_val]
    test_ids = eligible_ids[n_train + n_val:]

    print(f"Train IDs: {len(train_ids)}, Val IDs: {len(val_ids)}, Test IDs: {len(test_ids)}")
    return train_ids, val_ids, test_ids

def filter_pairs_by_ids(pairs_dict, keep_ids):
    return {pid: pairs_dict[pid] for pid in keep_ids}

# Perform the split using the binarized pairs (range [0,1])
eligible_ids = list(pairs_bin.keys())
train_ids, val_ids, test_ids = split_person_ids(eligible_ids)

# Create datasets for each split (use filtered binarized pairs)
BATCH_SIZE = 16
train_pairs = filter_pairs_by_ids(pairs_bin, train_ids)
val_pairs   = filter_pairs_by_ids(pairs_bin, val_ids)
test_pairs  = filter_pairs_by_ids(pairs_bin, test_ids)

train_ds = make_tf_dataset_from_pairs(train_pairs, batch_size=BATCH_SIZE, shuffle=True,  seed=SEED)
val_ds   = make_tf_dataset_from_pairs(val_pairs,   batch_size=BATCH_SIZE, shuffle=False, seed=SEED)
test_ds  = make_tf_dataset_from_pairs(test_pairs,  batch_size=BATCH_SIZE, shuffle=False, seed=SEED)

# Smoke test: take one batch from train_ds
for (x1, x2), y in train_ds.take(1):
    print(f"x1: {x1.shape}, x2: {x2.shape}, y: {y.shape}")
```

```python
# Minimal shared CNN base and pair head
def build_base(input_shape=(CANVAS_SIZE[0], CANVAS_SIZE[1], 1)):
    inputs = keras.Input(shape=input_shape)
    x = keras.layers.Conv2D(16, 3, padding='same', activation='relu')(inputs)
    x = keras.layers.MaxPool2D(2)(x)
    x = keras.layers.Conv2D(32, 3, padding='same', activation='relu')(x)
    x = keras.layers.MaxPool2D(2)(x)
    x = keras.layers.Conv2D(64, 3, padding='same', activation='relu')(x)
    x = keras.layers.GlobalAveragePooling2D()(x)
    x = keras.layers.Dense(128, activation='relu')(x)
    x = keras.layers.Dense(64, activation=None)(x)
    model = keras.Model(inputs, x, name='signature_base')
    return model

def build_siamese(input_shape=(CANVAS_SIZE[0], CANVAS_SIZE[1], 1)):
    base = build_base(input_shape)
    a = keras.Input(shape=input_shape)
    b = keras.Input(shape=input_shape)
    ea = base(a)
    eb = base(b)
    # L2 distance then a logistic head
    diff = keras.layers.Lambda(lambda t: tf.sqrt(tf.reduce_sum(tf.square(t[0]-t[1]), axis=1, keepdims=True)))([ea, eb])
    logit = keras.layers.Dense(1, activation=None)(diff)
    model = keras.Model([a,b], logit, name='siamese_logit')
    return model
```

```python
# Compile model with BCE-with-logits
model = build_siamese()
model.compile(optimizer=keras.optimizers.Adam(1e-3),
              loss=keras.losses.BinaryCrossentropy(from_logits=True),
              metrics=[keras.metrics.AUC(name='auc'), keras.metrics.BinaryAccuracy(name='acc')])
model.summary()
```

```python
EPOCHS = 12
history = model.fit(train_ds, validation_data=val_ds, epochs=EPOCHS)
```

```python
# 5) Before-vs-After training comparison (metrics + confusion matrix)
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

def evaluate_and_confusion(model_obj, dataset, title_prefix=""):
    # Metrics
    metrics = model_obj.evaluate(dataset, return_dict=True, verbose=0)
    # Collect labels
    y_true_list = []
    for _, y in dataset:
        y_true_list.append(tf.reshape(y, [-1]))
    y_true = tf.concat(y_true_list, axis=0)  # (N,)
    # Predictions (logits -> probs -> binary)
    logits = model_obj.predict(dataset, verbose=0)  # (N,1)
    logits = np.squeeze(logits, axis=-1)
    probs = 1.0 / (1.0 + np.exp(-logits))
    y_pred = (probs >= 0.5).astype(np.int32)
    y_true_np = tf.cast(y_true, tf.int32).numpy()
    cm = tf.math.confusion_matrix(y_true_np, y_pred, num_classes=2).numpy()
    # Print metrics
    print(f"{title_prefix} metrics:", {k: float(v) if hasattr(v, '__float__') else v for k, v in metrics.items()})
    # Plot confusion
    plt.figure(figsize=(4,4))
    plt.imshow(cm, cmap='Blues')
    plt.title(f'{title_prefix} Confusion Matrix (thr=0.5)')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.xticks([0,1],["gf","gg"])
    plt.yticks([0,1],["gf","gg"])
    for i in range(2):
        for j in range(2):
            plt.text(j, i, int(cm[i, j]), ha='center', va='center', color='black')
    plt.tight_layout()
    plt.show()

# Ensure required objects exist
if 'test_ds' not in globals():
    print("'test_ds' not found. Re-run dataset cells.")
else:
    # Build a fresh, untrained model with same architecture and compile
    model_untrained = build_siamese()
    model_untrained.compile(optimizer=keras.optimizers.Adam(1e-3),
                            loss=keras.losses.BinaryCrossentropy(from_logits=True),
                            metrics=[keras.metrics.AUC(name='auc'), keras.metrics.BinaryAccuracy(name='acc')])
    # Evaluate BEFORE training
    evaluate_and_confusion(model_untrained, test_ds, title_prefix="Before Training")
    # Evaluate AFTER training (current 'model')
    if 'model' in globals():
        evaluate_and_confusion(model, test_ds, title_prefix="After Training")
    else:
        print("'model' not found. Train the model to run the after-training comparison.")
```

```python
# 6) Before-vs-After t-SNE of pair embeddings (subset of test)
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
try:
    from sklearn.manifold import TSNE
except Exception as e:
    TSNE = None
    print("scikit-learn not available. Install it to run t-SNE.")

def compute_pair_features(base_model, dataset_subset):
    feats = []
    labels = []
    for (x1, x2), y in dataset_subset:
        e1 = base_model(x1, training=False)
        e2 = base_model(x2, training=False)
        d  = tf.abs(e1 - e2)
        feats.append(d)
        labels.append(tf.reshape(y, [-1]))
    if not feats:
        return None, None
    D = tf.concat(feats, axis=0).numpy()
    Y = tf.concat(labels, axis=0).numpy().astype(int)
    return D, Y

if 'test_ds' not in globals():
    print("'test_ds' not found. Re-run dataset cells.")
elif TSNE is None:
    print("t-SNE skipped: scikit-learn not available.")
else:
    # Prepare subset once to ensure identical pairs for both models
    subset = test_ds.unbatch().take(1000).batch(256)
    # Untrained base
    model_untrained = build_siamese()
    model_untrained.compile(optimizer=keras.optimizers.Adam(1e-3),
                            loss=keras.losses.BinaryCrossentropy(from_logits=True),
                            metrics=[keras.metrics.AUC(name='auc'), keras.metrics.BinaryAccuracy(name='acc')])
    base_untrained = model_untrained.get_layer('signature_base')
    D0, Y0 = compute_pair_features(base_untrained, subset)
    # Trained base
    if 'model' not in globals():
        print("'model' not found. Train the model for after-training t-SNE.")
    else:
        base_trained = model.get_layer('signature_base')
        # Recreate subset iterator (some environments exhaust iterables)
        subset2 = test_ds.unbatch().take(1000).batch(256)
        D1, Y1 = compute_pair_features(base_trained, subset2)
        if D0 is None or D1 is None:
            print("No features computed; ensure test_ds has data.")
        else:
            # Run separate t-SNEs for clarity
            tsne0 = TSNE(n_components=2, perplexity=30, init='pca', random_state=42)
            Z0 = tsne0.fit_transform(D0)
            tsne1 = TSNE(n_components=2, perplexity=30, init='pca', random_state=42)
            Z1 = tsne1.fit_transform(D1)
            colors = np.array(['tab:orange','tab:blue'])
            fig, axes = plt.subplots(1, 2, figsize=(12,5), sharex=False, sharey=False)
            axes[0].scatter(Z0[:,0], Z0[:,1], c=colors[Y0], s=10, alpha=0.7)
            axes[0].set_title('Before Training')
            axes[1].scatter(Z1[:,0], Z1[:,1], c=colors[Y1], s=10, alpha=0.7)
            axes[1].set_title('After Training')
            for ax in axes:
                ax.set_xticks([]); ax.set_yticks([])
            from matplotlib.lines import Line2D
            legend_elems = [
                Line2D([0],[0], marker='o', color='w', label='gf (0)', markerfacecolor='tab:orange', markersize=6),
                Line2D([0],[0], marker='o', color='w', label='gg (1)', markerfacecolor='tab:blue', markersize=6),
            ]
            fig.legend(handles=legend_elems, loc='upper right')
            fig.suptitle('t-SNE of |e1 - e2| (subset of test)')
            plt.tight_layout()
            plt.show()
```
