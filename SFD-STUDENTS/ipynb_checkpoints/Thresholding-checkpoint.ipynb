```python
def display_img(image):
    fig, ax = plt.subplots(1, 2, figsize=(10, 5))
    ax[0].imshow(image, cmap='gray')
    ax[0].axis('off')
    ax[0].set_title('Image')
    ax[1].hist(image.ravel(), bins=256, range=(0, 256))
    ax[1].set_title('Histogram')
    ax[1].set_xlabel('Pixel Intensity')
    ax[1].set_ylabel('Frequency (log scale)')
    ax[1].set_yscale('log')
    fig.tight_layout()
    plt.show()
    plt.close()
```

```python
def rotate(image, theta) -> np.ndarray:
    rotation_matrix = cv2.getRotationMatrix2D((image.shape[1]//2, image.shape[0]//2), theta, 1.0)
    rotated = cv2.warpAffine(image, M = rotation_matrix, dsize=(image.shape[1], image.shape[0]), borderValue=255)
    return rotated
```

```python
def compute_height(binary_img, theta):
    H, W = binary_img.shape
    M = cv2.getRotationMatrix2D((W // 2, H // 2), theta, 1.0)

    rotated = cv2.warpAffine(
        binary_img,
        M,
        (W, H),
        flags=cv2.INTER_NEAREST,
        borderValue=255
    )

    # assumes dark signature on bright background
    foreground = rotated < 128
    rows = np.any(foreground, axis=1)

    if not np.any(rows):
        return 0

    top = np.argmax(rows)
    bottom = len(rows) - np.argmax(rows[::-1])
    return bottom - top
```

```python
def optimal_angle(img):
    min, angle = 277, 0
    for i in range(-90,90):
        test = compute_height(img, i)
        if(test<= min): min, angle = test, i
    return min, angle

            

```
