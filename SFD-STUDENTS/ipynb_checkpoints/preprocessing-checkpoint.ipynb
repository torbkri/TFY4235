### Filter large images to reduce memory usage

```python

forg_dir = "signatures-dataset/full_forg"
org_dir = "signatures-dataset/full_org"

# Function to load images and extract metadata from filenames
def load_images_with_metadata(directory, real_bool, crop=None):
    """Load images and extract metadata (real/fake, person_id, image_id) from filenames."""
    data = []
    print(f"Loading images from directory: {directory}")
    print(f"Files in directory: {os.listdir(directory)}")  # Debug: List all files in the directory

    for filename in tqdm(os.listdir(directory), desc=f"Loading images from {directory}"):
        if filename.endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff')):
            img_path = os.path.join(directory, filename)
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

            if img is None:
                print(f"Warning: Unable to read image {img_path}")  # Debug: Check if image is read
                continue

            # Extract metadata from filename using regex
            match = re.match(r".*_(\d+)_(\d+)\..*", filename)
            if match:
                person_id = int(match.group(1))  # Extract person ID
                image_id = int(match.group(2))  # Extract image ID
                data.append({
                    "real": real_bool,
                    "person_id": person_id,
                    "image_id": image_id,
                    "image_path": img_path,
                    "image": img,  # Ensure the image is added to the DataFrame
                })
            else:
                print(f"Warning: Filename does not match pattern: {filename}")  # Debug: Check regex match

    print(f"Loaded {len(data)} images from {directory}")  # Debug: Check how many images were loaded
    return pd.DataFrame(data)

# Reload data into Pandas DataFrames to ensure the `image` column is populated
if 'forg_df' not in globals():
    forg_df = load_images_with_metadata(forg_dir, real_bool=False)

if 'org_df' not in globals():
    org_df = load_images_with_metadata(org_dir, real_bool=True)

# Combine the two DataFrames
all_data_df = pd.concat([org_df, forg_df], ignore_index=True)
print(all_data_df.head())  # Debug: Check the contents of the DataFrame
```

#### Crop images first

```python
def crop_img(img, padding=3):
    median_img = cv2.medianBlur(img, 5)
    curve_detector = np.array([[-1, -1, -1],
                                [-1,  8, -1],
                                [-1, -1, -1]])
    curve_img = cv2.filter2D(median_img, -1, curve_detector)
    # Add safety-padding to ensure that boundingRect does not cut off parts of the signature
    padded_img = cv2.copyMakeBorder(curve_img,3,3,3,3,borderType=cv2.BORDER_CONSTANT,value=0)
    _, thresh = cv2.threshold(padded_img, 1, 255, cv2.THRESH_OTSU)
    coords = cv2.findNonZero(thresh)
    x, y, w, h = cv2.boundingRect(coords)
    cropped_img = img[y-padding:y+h+padding, x-padding:x+w+padding]
    assert cropped_img.size > 0, "Cropped image has zero size"
    if not (cropped_img.shape[0] > 40 and cropped_img.shape[1] > 100):
        display_img(img, cropped_img)
        display_img(median_img, curve_img)
        display_img(padded_img, cropped_img)
        raise ValueError("Cropped image has invalid dimensions")
    return cropped_img

all_data_df = all_data_df.copy()
all_data_df['cropped_image'] = all_data_df['image'].apply(crop_img)
```

```python
def filter_large_images(df, max_height, max_width):
    """Filter out images with height > max_height or width > max_width."""
    df = df.copy()
    df['height'] = df['image'].apply(lambda x: x.shape[0])
    df['width'] = df['image'].apply(lambda x: x.shape[1])
    filtered_df = df[(df['height'] <= max_height) & (df['width'] <= max_width)]
    print(f"Filtered out {len(df) - len(filtered_df)} images exceeding {max_height}x{max_width} pixels.")
    print(f"Remaining images: {len(filtered_df)}")
    return filtered_df
```

```python
def filter_by_person_and_size(df, total_persons, signature_pairs_per_person, max_width=None, max_height=None, random_state=42):
    """Return exactly `total_persons` persons, each with exactly `signature_pairs_per_person` real and forged images."""
    df = df.copy()

    # Prefer cropped size; fall back to original
    df['height'] = df['cropped_image'].apply(lambda x: x.shape[0] if x is not None else None)
    df['width']  = df['cropped_image'].apply(lambda x: x.shape[1] if x is not None else None)
    missing = df['height'].isna()
    if missing.any():
        df.loc[missing, 'height'] = df.loc[missing, 'image'].apply(lambda x: x.shape[0])
        df.loc[missing, 'width']  = df.loc[missing, 'image'].apply(lambda x: x.shape[1])

    # Optional size upper bounds
    if max_height is not None:
        df = df[df['height'] <= max_height]
    if max_width is not None:
        df = df[df['width'] <= max_width]

    pairs = signature_pairs_per_person
    imgs_per_person = pairs * 2  # 4 real + 4 forged => 8 total

    # Keep only persons with enough balanced images
    counts = df.groupby(['person_id', 'real']).size().unstack(fill_value=0)
    eligible_persons = counts[(counts.get(True, 0) >= pairs) & (counts.get(False, 0) >= pairs)].index

    if len(eligible_persons) < total_persons:
        have = len(eligible_persons)
        raise ValueError(f"Not enough eligible persons with {pairs} real and {pairs} forged each: need {total_persons}, have {have}.")

    # Choose exactly total_persons persons (deterministic)
    rng = np.random.default_rng(random_state)
    selected_persons = rng.choice(eligible_persons, size=total_persons, replace=False)

    # Sample exactly `pairs` real and `pairs` forged per selected person
    selected_rows = []
    for pid in selected_persons:
        per_df = df[df['person_id'] == pid]
        real_df = per_df[per_df['real'] == True]
        forg_df = per_df[per_df['real'] == False]

        real_samp = real_df.sample(n=pairs, random_state=random_state)
        forg_samp = forg_df.sample(n=pairs, random_state=random_state)

        selected_rows.append(pd.concat([real_samp, forg_samp], ignore_index=True))

    final_df = pd.concat(selected_rows, ignore_index=True)

    # Sanity checks
    per_counts = final_df.groupby('person_id').size()
    assert len(per_counts) == total_persons, "Did not select exactly the requested number of persons."
    assert (per_counts == imgs_per_person).all(), "Did not select exactly the requested images per person."

    per_bal = final_df.groupby(['person_id', 'real']).size().unstack(fill_value=0)
    assert (per_bal[True] == pairs).all() and (per_bal[False] == pairs).all(), "Per-person real/forged balance is not correct."

    print(f"Filtered dataset to {len(final_df)} signatures: {total_persons} persons × {pairs} real + {pairs} forged (total {imgs_per_person}).")
    return final_df

# max_height, max_width = 9000, 571
# max_height, max_width = 348, 9000
factor = 1.08
max_height, max_width = int(348*factor), int(571*factor)
filtered_data_df = filter_large_images(all_data_df, max_height, max_width)

# Apply filtering by person, size, and total persons
total_persons = 32
signature_pairs_per_person = 4

final_data_df = filter_by_person_and_size(filtered_data_df, total_persons, signature_pairs_per_person,
                max_width=None, max_height=None, random_state=42)
```

#### Pad and center small signatures to standardize resolution

```python
# Invert and pad all images to the same size: max_height, max_width.
def finalize_images(df, target_height, target_width, binarize=True):
    """Invert and pad all images in the DataFrame to the target height and width."""
    def finalize_image(img):
        if binarize:
            _, img = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
        else:
            img = cv2.bitwise_not(img)
        h, w = img.shape
        pad_height = target_height - h
        pad_width = target_width - w
        top = pad_height // 2
        bottom = pad_height - top
        left = pad_width // 2
        right = pad_width - left
        padded_img = cv2.copyMakeBorder(img, top, bottom, left, right, cv2.BORDER_CONSTANT, value=0)
        return padded_img

    df = df.copy()
    df['padded_image'] = df['cropped_image'].apply(finalize_image)
    assert all(df['padded_image'].apply(lambda x: x.shape == (target_height, target_width))), "Not all images were padded correctly."
    return df

final_data_df_binarized = finalize_images(final_data_df, max_height, max_width)
final_data_df = finalize_images(final_data_df, max_height, max_width, binarize=False)
```

#### Save filtered signatures

```python
import os
from PIL import Image

# Create directories for filtered images
filtered_org_dir = "signatures-dataset/filtered_org"
filtered_forg_dir = "signatures-dataset/filtered_forg"
filtered_org_dir_binarized = "signatures-dataset/filtered_org_binarized"
filtered_forg_dir_binarized = "signatures-dataset/filtered_forg_binarized"
os.makedirs(filtered_org_dir, exist_ok=True)
os.makedirs(filtered_forg_dir, exist_ok=True)
os.makedirs(filtered_org_dir_binarized, exist_ok=True)
os.makedirs(filtered_forg_dir_binarized, exist_ok=True)

final_data_df['image_path_filtered'] = final_data_df.apply(
    lambda row: os.path.join(
        filtered_org_dir if row['real'] else filtered_forg_dir,
        f"{'original' if row['real'] else 'forgeries'}_{row['person_id']}_{row['image_id']}.png"
    ),
    axis=1
)
final_data_df_binarized['image_path_filtered'] = final_data_df_binarized.apply(
    lambda row: os.path.join(
        filtered_org_dir_binarized if row['real'] else filtered_forg_dir_binarized,
        f"{'original' if row['real'] else 'forgeries'}_{row['person_id']}_{row['image_id']}.png"
    ),
    axis=1
)

# Save filtered images to the appropriate folders
def save_filtered_images(df, org_dir, forg_dir):
    """Save filtered images to directories with the same naming format."""
    # assert that there are 4 real and 4 forged images per person
    imgs_per_person = df.groupby(['person_id', 'real']).size().unstack(fill_value=0)
    assert (imgs_per_person[True] == 4).all() and (imgs_per_person[False] == 4).all(), \
            "Each person must have exactly 4 real and 4 forged images."
    for _, row in df.iterrows():
        person_id = row['person_id']
        image_id = row['image_id']
        real = row['real']
        image = row['padded_image']
        
        # Determine the output directory and filename
        if real:
            output_dir = org_dir
            filename = f"original_{person_id}_{image_id}.png"
        else:
            output_dir = forg_dir
            filename = f"forgeries_{person_id}_{image_id}.png"
        
        # Save the image
        output_path = os.path.join(output_dir, filename)
        Image.fromarray(image).save(output_path)

# Save the filtered images
save_filtered_images(final_data_df, filtered_org_dir, filtered_forg_dir)
print(f"Filtered images saved to '{filtered_org_dir}' and '{filtered_forg_dir}'.")
save_filtered_images(final_data_df_binarized, filtered_org_dir_binarized, filtered_forg_dir_binarized)
print(f"Filtered binarized images saved to '{filtered_org_dir_binarized}' and '{filtered_forg_dir_binarized}'.")

# Save the filtered dataframe to CSV including only padded images
final_data_df = final_data_df.drop(columns=['image', 'padded_image', 'cropped_image', 'height', 'width'])
final_data_df.to_csv("signatures-dataset/filtered_dataframe.csv", index=False)
print("Filtered dataframe saved to 'filtered_dataframe.csv'.")
final_data_df_binarized = final_data_df_binarized.drop(columns=['image', 'padded_image', 'cropped_image', 'height', 'width'])
final_data_df_binarized.to_csv("signatures-dataset/filtered_dataframe_binarized.csv", index=False)
print("Filtered binarized dataframe saved to 'filtered_dataframe_binarized.csv'.")
```
