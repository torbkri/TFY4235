```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def get_filter(shape, radius):
    H, W = shape
    y = np.arange(H)
    x = np.arange(W)
    
    yy, xx = np.meshgrid(y, x, indexing='ij')
    
    cy = H // 2
    cx = W // 2
    
    return (yy - cy)**2 + (xx - cx)**2 <= radius**2
```

```python
img = cv2.imread("/Users/torbjornkringeland/Desktop/NTNU/6/TFY4235/SFD-STUDENTS/exampleSignature.png", cv2.IMREAD_GRAYSCALE)
img = np.load("/Users/torbjornkringeland/Desktop/NTNU/6/TFY4235/SFD-STUDENTS/exampleSignature_gaussian.npy")
```

```python
ft_img = np.fft.fft2(img)
# ft_img = np.abs(ft_img)

ft_img.max()
centered = np.fft.fftshift(ft_img)


# result =np.abs(np.fft.ifft2(filtered))
```

```python
mask = get_filter(ft_img.shape, 50)

filtered = np.zeros(centered.shape,dtype= complex)
for y in range(filtered.shape[0]):
    for x in range(filtered.shape[1]):
        filtered[y][x] = mask[y][x] * (centered[y][x])
```

```python
result = np.fft.ifft2(np.fft.ifftshift(filtered))
plt.imshow(np.abs(result))
```
